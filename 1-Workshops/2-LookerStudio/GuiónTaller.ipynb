## Business Intelligence: ¿Qué es?

La inteligencia empresarial o de mercado es el uso de estrategias y herramientas que sirven para transformar información en conocimiento, con el objetivo de mejorar, agilizar y optimizar la toma de decisiones en una empresa así como solucionar conflictos en tiempo real.

Trabajan con enormes cantidades de datos en la nube traducidos en métricas (KPIs) de los principales negocios o áreas de una compañía.

### Herramientas en el mercado
#### Power BI

Herramienta de Microsoft para BI. Es la lider de mercado sin duda, en cuestión de 5 años ha pasado de ser una herramienta secundaria a saber sacar partido a su servicio cloud, Azure. Es actualizado mensualmente y ha pasado de ser limitado en opciones a tener mucha potencia tanto en procesado de datos, preparado de datos y visualización.

- **Power BI Desktop**: aplicación de escritorio (solo Windows) para transformar, visualizar datos y crear informes de los mismos.
- **Power BI Service**: servicio online que permite publicar informes y configurar la actualización automática para la toma de decisiones.
- **Power BI Mobile**: aplicación móvil compatible con iOS y Android para tener acceso a la visualización de informes, no edición.

Puedes consultar toda la documentación gratuita de Power BI [aquí](https://learn.microsoft.com/es-es/power-bi/guidance/) y también está disponible un foro es español para dudas en Telegram: https://web.telegram.org/#/im?p=@powerbiespanol

#### Qlik

- **QlikView (básica)**: Versión en deprecación, pero presente aún en muchos negocios, basada en un cliente instalable y despliegue web
- **Qlik Sense (profesional completa)**: Versión web, *responsive* con muy buen rendimiento y amplia versatilidad
- **Qlik Sense Mobile**

![imagen](images/qlikcomparison.png)

#### Looker Studio

Herramienta online totalmente gratuita de Google para transformar los datos en informes y paneles de visualización fáciles de consultar y compartir. Es una herramienta 100% web, con amplia versatilidad para conectar con datos y fácil uso. También es la más limitada de todas a la hora de crear dashboards complejos.

- **Looker Studio**: Informes gratuítos de hasta 1 GB, genial para aprender BI y hacer dashboards simples pero efectivos
- **Looker Studio Pro**: Versión con mejor capacidad, deployment y capacidad orientada para consumo masivo de visitas.

https://lookerstudio.google.com/u/0/navigation/reporting

#### Tableau

Segunda en volumen de mercado, es una herramienta con unos orígenes parecidos a Qlik, pero que ha sabido mantener su esencia con el tiempo. Actualmente la compañía pertenece a Salesforce, software de CRM (Customer Resource Management). Es robusta, muy potente para consumo masivo de datos, pero limitada en el modelo de datos.

https://www.tableau.com/es-es/products/trial

### Storytelling
![imagen](images/storytellingpipeline.png)

Cómo crear una presentación efectiva, es decir, maximizar nuestra capacidad de impacto e influencia.

**SIN STORYTELLING**

- somos olvidados.
- los datos son difíciles de memorizar.
- quedamos a la merced del detalle.

**¿Qué es una historia y por qué a las personas nos encantan?**

- Crea emoción y empatía.
- Hace que seamos memorables.
- Aumenta el recuerdo de los detalles.
- Evoca la imaginación en la audiencia, conectando con el subconsciente.

**Pilares**

- Mensaje o mensajes (objetivo)
    
    ¿qué quieres que **PIENSE** tu audiencia?
    
    ¿qué quieres que **SIENTA** tu audiencia?
    
    ¿qué quieres que **HAGA** tu audiencia?
    
- Guión (contenido / estructura): comienzo, nudo, desenlace.
    - **Personajes** (protagonistas) que sirven para empatizar con la audiencia.
    - **Método** **LOCK,** 4 fases:
        
        Lead (protagonista → audiencia)
        
        Objetive (mensaje) 
        
        Confrontation (dificultades)
        
        Knock End (la gran batalla final)
        
    - **Apertura y Cierre**
        
        El hilo conductor con un inicio impactante que conecte con la audiencia y enganche hasta el desenlace.
        
- Escenificación
    
    - Formato.
    
    - Lenguaje sensorial.
    
    - El poder de la voz: entonación, volumen, silencios, etc.
    
    - Interacción con la audiencia.
    
- Ensayo
    
    - Grábate, cronométrate, pide feedback, pule los detalles.

### Práctica Fútbol Americano
En este ejercicio se espera que practiques lo visto durante la clase. Los datos que utilizarás
para este fin son los del archivo `nfl_football_team_stats.csv` que encontrarás en la
carpeta data.

Este archivo contiene estadísticas de fútbol americano, sobre los partidos jugados por cada
equipo. No necesitas ser un experto en este deporte para crear gráficos que resuman
dichos eventos pero si quieres aprender un poco más, te dejo por aquí un par de enlaces:

- https://www.youtube.com/watch?v=3t6hM5tRlfA
- https://en.wikipedia.org/wiki/American_football_strategy

Parte de tu trabajo es y será familiarizarte con aquellos datos con los que tengas que
trabajar.

#### 1. Partidos disputados por año:

Sacar el total de partidos disputados por año (2010-2017) en un gráfico de barras. ¿Qué conclusión podemos sacar de esta gráfica?
- Identificar una variable (dimensión o hecho) para contabilizar los partidos
- Identificar la variable de año o calcular una a partir de otra

#### 2. Proporción de partidos ganados por cada equipo y año

Ahora queremos visualizar la proporción de partidos ganados y perdidos para cada
equipo por año. Para ello vamos a intentar una gráfica algo compleja pero a con mucha información. Tenemos que extraer de alguna manera los partidos ganados y pintarlos en una tabla para los equipos y para los años, columnas años y filas equipos. ¿Cómo podemos hacerlo?

#### 3. Averigua si jugar en casa o fuera tiene un impacto en que el equipo gane.

Para ello, queremos saber si se gana más en casa o fuera de ella. Podemos intentarlo con un gráfico de barras enfrentadas para ambas hipótesis.

#### 4. Podemos determinar la eficiencia de ofensa y defensa por el número de puntos por equipo y por oponente, respectivamente.

Necesitamos ver en dos ejes, los puntos de media que saca cada equipo cuando juega en casa y cuando juega fuera pintados en el mismo sitio. Cada punto debe ser un equipo y así veremos como se coloca cada uno sobre estos dos scores. ¿Quién es el mejor?

#### 5. Lo siguiente es echar un vistazo a las estadísticas de pase (Passing) del año 2016

Para ello tenemos que evaluar los pases realizados, los intentados y también sacar un promedio de pasos completados. Si puede ser todo en la misma gráfica mejor. Hazlo con gráficos de barras en horizontal.

- Extra: ¿Se te ocurre otra manera de poder representar el promedio de los pases para todos los equipos de manera diferente? Propon uno

#### 6. Ahora es el turno de las estadísticas de defensa, específicamente del average number of sacks, interceptions y safeties para cada equipo, cuando gana.

Para ello lo mejor es usar de nuevo un gráfico de barras horizontal con cada uno de los kpis de Defensa (interceptions, Sacks y Sefeties)

#### 7. Crea una última visualización que de un vistazo nos permita ver lo bien que los equipos defienden en base a la distribución de puntos que reciben

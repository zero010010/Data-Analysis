Datos del Problema: 
* 740 casas 1 / 1 contador por casa/ 1 registro por hora 

* en la base de datos las filas y columnas no estan numeradas 

* maximo 16 registros en columna, pero puede haber filas infinitas

* en las filas hay que guardar 24 registros, y solo guardan 16.

* cada 24h  irían guardando los datos de cada vecino de forma consecutiva y ordenada. 

* Primero el vecino1, luego el vecino2, luego el vecino3 y así hasta el vecino740. 

* Hay que identificar a que día corresponde cada conjunto de registros

* el primer día de datos que corresponde al 16 de enero de 2004.

* no hay ningun registro nulo en todo el histórico de datos del sistema


Preguntas del problema:

* Crear una matriz que cada fila corresponde a un día del año, y cada columna corresponde a un cliente. 
* Cada celda se encontrará una lista de los registros que marca el contador. 
* Crear otra matriz, en la que en vez de tener los registros, tengamos la 
    media de estos registros por cliente y día. 
* Sabiendo que la energía cuesta 0.25€ el KW/h, crear una nueva matriz en la que por cada cliente y día, 
tengamos el coste total de cada cliente, para tener la facturación diaria de estos.


```python
import numpy as np
import pandas as pd
# crear una matriz que cada fila corresponde a un día del año, y cada columna corresponde a un cliente.

# Crear una matriz de ceros con 365 filas y 10 columnas
matriz = np.zeros((365, 10))

# Iterar sobre cada fila y columna y asignar un valor a cada elemento de la matriz
for i in range(365):
    for j in range(10):
        matriz[i][j] = i + j

# Convertir la matriz en un DataFrame de Pandas
df = pd.DataFrame(matriz)

# Asignar nombres de columna a cada columna del DataFrame
df.columns = ['vecino ' + str(i) for i in range(1, 11)]

# Asignar nombres de fila a cada fila del DataFrame
df.index = pd.date_range(start='2023-01-01', end='2023-12-31')

# Imprimir el DataFrame
print(df)
```

```python
import numpy as np
import pandas as pd

# Crear una matriz de ceros con 365 filas y 10 columnas
matriz = np.zeros((365, 10))

# Iterar sobre cada fila y columna y asignar un valor a cada elemento de la matriz
for i in range(365):
    for j in range(10):
        matriz[i][j] = i + j

# Convertir la matriz en un DataFrame de Pandas
df = pd.DataFrame(matriz)

# Asignar nombres de columna a cada columna del DataFrame
df.columns = ['Cliente ' + str(i) for i in range(1, 11)]

# Asignar nombres de fila a cada fila del DataFrame
df.index = pd.date_range(start='2023-01-01', end='2023-12-31')

# Calcular el coste total de cada cliente por día y asignarlo a una nueva matriz
matriz_coste = matriz * 0.25

# Convertir la matriz de costes en un DataFrame de Pandas
df_coste = pd.DataFrame(matriz_coste)

# Asignar nombres de columna a cada columna del DataFrame
df_coste.columns = ['Cliente ' + str(i) for i in range(1, 11)]

# Asignar nombres de fila a cada fila del DataFrame
df_coste.index = pd.date_range(start='2023-01-01', end='2023-12-31')

# Imprimir el DataFrame
print(df_coste)
```

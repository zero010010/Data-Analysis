Rellenar una matriz numpy vacia de tamaño 365,740 con una lista en cada celda DE 24 REGISTROS extraidos de la matriz data:
np.random.seed(2023)
data = np.random.randint(low=12,high=150,size=(999000,16))

pero teniendo en cuenta que los 24 registros que estaran dentro de cada celda no están en la misma fila de la matriz inicial, si no que estaran formados por, para el primer vecino, la primera fila y media de la segunda de la matriz data; las del siguiente vecino, la media segunda fila que falta y la tercera entera; para el tercer vecino, la 4º fila de data y la mitad de la quinta de data... y asi para todos los vecinos en todos los dias

PARA EL PROBLEMA PRINCIPAL, QUE ERA NO PODER SELECCIONAR LA PRIMERA FILA DE DATA PARA EL PRIMER VECINO, LA SEGUNDA PARA EL SEGUNDO QUE ERA LA FORMA FACIL, lo hemos solucionado asi:
- Hemos creado un TRIPLE BUCLE, primer bucle recorre filas, segundo columnas y el tercero LOS REGISTROS QUE NECESITAMOS(24) que localizamos a traves de INDICES, y realizamos las operaciones necesarias para que los índices se actualicen correctamente para cada cliente (sumamos a indice de inicio los 24 registros para cada nuevo cliente, tenemos en cuenta el tamaño de la matriz para hacer que la selección sea cíclica (cuando termine fila, que pase a la siguiente) y transformamos los índices planos en bidimensionales.) Los 24 registros de cada cliente Y EN CADA CELDA se almacenan en una lista vacía llamada DATOS_CLIENTE y posteriormente la transformamos en matriz numpy y la almacenamos en la celda de la matriz resultante.

```python
import numpy as np

# Generar la matriz de datos original
np.random.seed(2023)
data = np.random.randint(low=12, high=150, size=(999000, 16))

# Dimensiones de la matriz resultante
filas_resultado, columnas_resultado = 365, 740

# Número de registros por cliente
registros_por_cliente = 24

# Crear la matriz EN LA QUE SE INCLUIRAN LOS DATOS DE LA ORIGINAL,POR CELDAS
matriz_resultado = np.empty((filas_resultado, columnas_resultado), dtype=object) # la creamos vacía para rellenarla luego

# Índice de inicio en la matriz de datos
indice_inicio = 0

# Bucle para recorrer las filas
for i in range(filas_resultado):
    # Bucle para recorrer las columnas
    for j in range(columnas_resultado):
        # Bucle para seleccionar los 24 registros consecutivos (EL MAS IMPORTANTE)
        datos_cliente = [] # LISTA VACIA PARA INTRODUCIR LOS DATOS QUE SE REQUIERE
        for k in range(registros_por_cliente):
            # Calcular el índice actual teniendo en cuenta el final de cada fila
            indice_actual = (indice_inicio + k) % data.size #  Si el índice calculado supera el tamaño total de la matriz, el operador % 
            # permite la seleccion cíclica de los registros.

            # Obtener la fila y la columna correspondientes(para convertir un índice plano (indice_actual) 
            # en un par de índices bidimensionales (fila_actual y columna_actual))
            fila_actual = indice_actual // data.shape[1]
            columna_actual = indice_actual % data.shape[1]

            # Agregar el dato actual a la lista
            datos_cliente.append(data[fila_actual, columna_actual])

        # Almacenar los datos en la celda correspondiente de la matriz resultante
        matriz_resultado[i, j] = np.array(datos_cliente)

        # Actualizar el índice de inicio para el siguiente cliente
        indice_inicio += registros_por_cliente

# La matriz_resultante ahora contiene los conjuntos de registros necesarios para cada cliente.
print(matriz_resultado)
print(matriz_resultado.shape)

```

```python
# Ahora, vamos a buscar la media de los arrays de cada celda:

# Crear una matriz para almacenar las medias 
medias_por_celda = np.empty_like(matriz_resultado, dtype=float) # mejor que np.zeros, porque vamos a sobreescribir los valores.

# Bucle para recorrer las filas
for i in range(filas_resultado):
    # Bucle para recorrer las columnas
    for j in range(columnas_resultado):
        # Calcular la media de los datos en la celda actual
        media_celda = np.mean(matriz_resultado[i, j])

        # Almacenar la media en la matriz de medias por celda
        medias_por_celda[i, j] = media_celda

# Ahora, medias_por_celda contiene las medias de cada array en cada celda de la matriz resultante
print(medias_por_celda)
```

```python
# Y AHORA, EL COSTE TOTAL POR DIA
# Costo por kilovatio
costo_por_kw = 0.25

# Crear una matriz para almacenar los costos totales de cada celda
costos_totales_por_celda = np.empty_like(matriz_resultado, dtype=float)

# Bucle para recorrer las filas
for i in range(filas_resultado):
    # Bucle para recorrer las columnas
    for j in range(columnas_resultado):
        # Sumar todos los valores en la celda y multiplicar por el costo por kilovatio
        costos_totales_por_celda[i, j] = np.sum(matriz_resultado[i, j]) * costo_por_kw

# Ahora, costos_totales_por_celda contiene los costos totales de cada celda en la matriz resultante
print(costos_totales_por_celda)
```
